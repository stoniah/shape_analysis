```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import numpy as np
```

```python
df = pd.read_excel(Path('Strain_with_trajectory_features.xlsx'))
df = df.dropna(axis=1)
```

```python
plt.figure(figsize=(15,10))
corr = df.corr()
features = corr.columns[((np.abs(corr) < 1) & (np.abs(corr) >= 0.5)).any()]
sns.heatmap(df[features].corr(), vmin=-1, vmax=1, cmap='coolwarm')
```
