```python
def f(x, roots, a0=1):
    """returns value of poly a0(x-x1)...(x-xn)"""
    return a0*np.product(np.subtract(x, roots),axis=1)

def f_lastprime(x, roots, a0=1):
    n = len(roots)
    n_fact = np.math.factorial(n-1)
    diff = np.subtract(n*x, np.sum(roots))
    return a0*n_fact*diff


n_roots = 100
# roots = np.repeat(0, n_roots)
roots = np.array([*np.repeat(0,n_roots-1), 1])
print(roots)
xlim = [-2,2]
xs = np.linspace(*xlim,10000)[:,np.newaxis]
ys = f(xs, roots)
print(np.subtract(xs, roots).shape)
# print(f(xs, roots))
plt.plot(xs, ys, zorder=1)
plt.plot(xs, f_lastprime(xs, roots))
# ylim = plt.ylim()
y_min=1e-6
ylim = [-y_min,y_min]
plt.plot(xlim, [0,0], color='black', linewidth=0.5, zorder=0)
plt.plot([0,0], ylim, color='black', linewidth=0.5, zorder=0)
plt.xlim(xlim)
plt.ylim(ylim)
# plt.gca().set_aspect('equal')
```

```python
def f(x):
    return 2*x-4+1/(x-1)**2

def f_prime(x):
    return 2-2/(x-1)**3

def step(x):
    return x - f(x)/f_prime(x)

xs = [0]
for i in range(10):
    xs.append(step(xs[-1]))
xs = np.array(xs)
plt.plot(xs, f(xs), marker='o')

points = np.linspace(-5,5)
plt.plot(points, f(points))
plt.ylim(-10,10)
ylim = plt.ylim()
xlim = plt.xlim()
plt.plot(xlim, [0,0], color='black', linewidth=0.5, zorder=0)
plt.plot([0,0], ylim, color='black', linewidth=0.5, zorder=0)
plt.xlim(xlim)
plt.ylim(ylim)
# plt.gca().set_aspect('equal')
```

```python
V = 1

def f(x):
    return V / x + np.pi * x**2 - V**(2/3)*np.pi**(1/3)*(2**(1/3)+2**(-2/3))*1.1

def f_prime(x):
    return -V/x**2+2*np.pi*x

def step(x):
    return x - f(x)/f_prime(x)

xs = [0.1]
for i in range(10):
    xs.append(step(xs[-1]))
xs = np.array(xs)
plt.plot(xs, f(xs), marker='o')

points = np.linspace(0.001,5,1000)
plt.plot(points, f(points))
plt.ylim(-1,10)
ylim = plt.ylim()
xlim = plt.xlim()
plt.plot(xlim, [0,0], color='black', linewidth=0.5, zorder=0)
plt.plot([0,0], ylim, color='black', linewidth=0.5, zorder=0)
plt.xlim(xlim)
plt.ylim(ylim)
# plt.gca().set_aspect('equal')
```
